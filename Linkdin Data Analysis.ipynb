```python
##
def bar_plot(dfname, colname, ncount=500):
    
    # data manipulation
    import pandas as pd
    data = pd.DataFrame(dfname)
    first_company_ = data[f"{colname}"].value_counts().head(ncount)
    first_company = first_company_.head(ncount)

    df = pd.DataFrame(first_company)
    dff = pd.DataFrame(df.reset_index())
    dff.rename(columns={'index': f"{colname}", f"{colname}": "freq"}, inplace=True)
    
    # creating barplot
    import matplotlib as mpl
    import matplotlib.pyplot as plt
    import seaborn as sns
    fig = plt.figure()
    fig.set_figwidth(7)
    fig.set_figheight(7)
    
    title = f"bar plot of {colname}"
    sns.barplot(data = dff, color="lightblue", x ="freq", y =f"{colname}").set(title=title)
    
    return 


bar_plot(data_1, "0.company",15)

    
```

#### - The company that the most of the people get their first job is Scotiabank. Then the others companies are CIBC ,TD, RBC etc.

```python
import pandas as pd
# stay date in a company
def stay_date(dfname, startDate, endDate, com):
    dataframe = pd.DataFrame(dfname)
    date = dataframe.dropna()
    date[f'stay_date_{com}'] = (date[f'{endDate}'] - date[f'{startDate}']).dt.days

    return date
```
